```python
from uright.inkutils import json2array
import numpy as np
user_id = 74

ink_data = {}

def interpolate_penups(traj, penups, num=3):
    '''
    traj can be nx3 (y,dx,dy) or nx2, the returned array is nx3.
    '''
    offset = 0
    for i in penups:
        if i + offset != len(traj)-1:
            a = np.linspace(0,1,num+2)[1:-1]
            intermediate = np.outer(1-a, traj[i+offset]) + np.outer(a, traj[i+offset+1])
            traj = np.vstack([traj[:offset+i+1,:], intermediate, traj[offset+i+1:,:]])    
            offset += num

    dif = np.diff(traj[:,:2], axis=0)
    dif_norm = np.sqrt(dif[:,0]**2+dif[:,1]**2)
    # dif_norm might be 0 if consecutive positions are the same
    dif[:,0] = dif[:,0]/(dif_norm + np.finfo(float).eps)
    dif[:,1] = dif[:,1]/(dif_norm + np.finfo(float).eps)
    ret = np.hstack([traj[:,:2], np.vstack([[0, 0], dif])])
    assert not np.isnan(np.sum(ret)), dif
    return ret

def json2array(ink, version):
    traj = np.array([[p['x'], p['y'], p['penup']]for p in ink['points']])
    penups = np.nonzero(traj[:,2] == 1)[0]
    traj_array = interpolate_penups(traj[:, :2], penups)
    return traj_array

for label,_,_,ink in all_user_data[user_id]:
    array_ink = json2array(ink, version='uright3')
    # remove nan
#     to_keep_idx = (array_ink[:,4] != 1)
#     array_ink = array_ink[to_keep_idx,0:4]

    array_ink = array_ink[:,0:4]
    
    # remove too short sample
    if array_ink.shape[0] > 8:
        ink_data.setdefault(label, []).append(array_ink)
```

```python
def o(c):
    return ord(c) - ord('a')

def cha(i):
    return chr(i + ord('a'))

import string

def gen_bigram(filenames):
    bigram = np.ones((26,26))
    for filename in filenames:
        f = open(filename,'r')
        for line in f:
            line = line.strip().lower()
            for i in range(len(line)-1):
                if (line[i] in string.ascii_lowercase and
                    line[i+1] in string.ascii_lowercase):
                    bigram[ord(line[i])-ord('a'), ord(line[i+1])-ord('a')] += 1
    for i in range(26):
        bigram[i,:] = bigram[i,:] / bigram[i,:].sum()
    return bigram
        
bigram = gen_bigram(['twain-adventures-27.txt','phrack.29.phk'])        
# np.savetxt('bigram', bigram)
```

```python
import random
from matplotlib import pyplot as plt

from matplotlib.patches import Circle
from matplotlib.lines import Line2D
from matplotlib.patches import Rectangle

def display_prototype(prot, fig_width=100, row_height=100):
#     fig = plt.figure(figsize=(fig_width, row_height))
    fig = plt.figure()
    ax = fig.add_subplot(1, 1, 0)
    draw_ink(ax, prot, show_order=True)
    ax.set_ylim(2.5,-2.5)
    ax.set_xlim(prot[:,0].min()-2, prot[:,0].max()+2)    
#     ax.set_title("Prototype for %s [%d]"%(prot.label,prot.num_obs))
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
    return fig

def display_prototype_seq(prot, seq, fig_width=100, row_height=100):
#     fig = plt.figure(figsize=(fig_width, row_height))
    fig = plt.figure()
    ax = fig.add_subplot(1, 1, 0)
    draw_ink_seq(ax, prot, seq, show_order=True)
    ax.set_ylim(2.5,-2.5)
    ax.set_xlim(prot[:,0].min()-2, prot[:,0].max()+2)    
#     ax.set_title("Prototype for %s [%d]"%(prot.label,prot.num_obs))
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
    return fig

def draw_ink_seq(ax, ink,  seq, node_size=0.1, penup_color='#0000FF',
             pendown_color='#00FF00', node_colors=None,
             show_order=True):
    
#     letter_seq = []
#     stages_seq = []
    label_seq = []
    for i in range(len(seq)/2):
#         letter_seq.append()
#         stages_seq.append(i+1)
        label_seq.append(seq[i*2:(i+1)*2])
    
    current_stroke_id = 0
    for i in range(ink.shape[0]):

        e = Circle(xy=(ink[i,0],ink[i,1]),radius=node_size, alpha=0.5)
        ax.add_artist(e)

        # pen-up/pen-down
        e.set_color(pendown_color)
        e.set_linewidth(5.0)
        current_stroke_id += 1

        if show_order:
            ax.text(ink[i,0]+node_size,ink[i,1],"[%d]%s"%(current_stroke_id,label_seq[i]))
            if label_seq[i][0] == '-':
                e.set_color(penup_color)
            
#         elif i == ink.shape[0]-1 or ink[i+1,4] > 0:
#             e.set_color(penup_color)
#             e.set_linewidth(5.0)

        if node_colors is not None:
            e.set_facecolor(node_colors[i])

        # draw arrow
        if (i < ink.shape[0]-1):
            dx = ink[i,2]
            dy = ink[i,3]
#             z = np.sqrt(dx*dx+dy*dy)
#             dx = dx / max(z,1e-5)
#             dy = dy / max(z,1e-5)
            if abs(dx) > 0 or abs(dy) > 0:
                ax.arrow(ink[i,0]-0.5*node_size*dx,
                         ink[i,1]-0.5*node_size*dy,
                         node_size*dx, node_size*dy,
                         fc="k", ec="k", alpha=0.5, width=0.007,
                         head_width=0.03, head_length=0.02,
                         length_includes_head=True)

    ax.axis('equal')


def draw_ink(ax, ink, node_size=0.1, penup_color='#0000FF',
             pendown_color='#00FF00', node_colors=None,
             show_order=False):
    current_stroke_id = 0
    for i in range(ink.shape[0]):

        e = Circle(xy=(ink[i,0],ink[i,1]),radius=node_size, alpha=0.5)
        ax.add_artist(e)

        # pen-up/pen-down
        e.set_color(pendown_color)
        e.set_linewidth(5.0)
        current_stroke_id += 1

        if show_order:
            ax.text(ink[i,0]+node_size,ink[i,1],"[%d]"%current_stroke_id)

#         elif i == ink.shape[0]-1 or ink[i+1,4] > 0:
#             e.set_color(penup_color)
#             e.set_linewidth(5.0)

        if node_colors is not None:
            e.set_facecolor(node_colors[i])

        # draw arrow
#         if (i <= ink.shape[0]-1):
        dx = ink[i,2]
        dy = ink[i,3]
#             z = np.sqrt(dx*dx+dy*dy)
#             dx = dx / max(z,1e-5)
#             dy = dy / max(z,1e-5)
        if abs(dx) > 0 or abs(dy) > 0:
            ax.arrow(ink[i,0]-0.5*node_size*dx,
                     ink[i,1]-0.5*node_size*dy,
                     node_size*dx, node_size*dy,
                     fc="k", ec="k", alpha=0.5, width=0.007,
                     head_width=0.03, head_length=0.02,
                     length_includes_head=True)

    ax.axis('equal')

def interpolate_letters(word, spacing=0.8):
    penups = []
    offset = 0
    counter = 0
    for i, c in enumerate(word):
#         chosen_prototype = ink_data[o(c)][0]
        chosen_prototype = random.choice(ink_data[o(c)])
        counter += chosen_prototype.shape[0]
        if i==0:
            ink = chosen_prototype
        else:
            offset_prototype = np.copy(chosen_prototype)
            offset_prototype[:,0] = offset_prototype[:,0]+ offset 
            ink = np.vstack([ink, offset_prototype])
        
        penups.append(counter-1)
        offset += chosen_prototype[:,0].max() + spacing
    
    ink = interpolate_penups(ink, penups)
#     print ink.min(axis=0)[0], ink.max(axis=0)[0]
    fig = display_prototype(ink)
#     plt.show()
    return ink, fig
```
